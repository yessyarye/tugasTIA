```python
import pandas as pd
import csv
import glob
from sklearn.model_selection import cross_val_score, RepeatedStratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import StackingClassifier, BaggingClassifier, AdaBoostClassifier
from sklearn.impute import SimpleImputer
from numpy import mean, std
from matplotlib import pyplot as plt

# Fungsi untuk mendeteksi pemisah otomatis
def detect_delimiter(file_path):
    with open(file_path, 'r') as file:
        sample = file.read(1024)  # Membaca sampel file
        delimiter = csv.Sniffer().sniff(sample).delimiter
    return delimiter

# Fungsi untuk membaca CSV dengan pemisah yang terdeteksi
def read_csv_with_detected_delimiter(file_path):
    delimiter = detect_delimiter(file_path)
    data = pd.read_csv(file_path, sep=delimiter)
    return data

# Fungsi untuk membuat stacking model
def get_stacking():
    level0 = list()
    level0.append(('lr', LogisticRegression(max_iter=1000)))
    level0.append(('knn', KNeighborsClassifier()))
    level0.append(('cart', DecisionTreeClassifier()))
    level0.append(('svm', SVC(probability=True)))
    level0.append(('bayes', GaussianNB()))
    level1 = LogisticRegression(max_iter=1000)
    model = StackingClassifier(estimators=level0, final_estimator=level1, cv=5)
    return model

# Fungsi untuk mendapatkan model ensemble
def get_models():
    models = dict()
    models['LR'] = LogisticRegression(max_iter=1000)
    models['KNN'] = KNeighborsClassifier()
    models['Decision Tree'] = DecisionTreeClassifier()
    models['SVM'] = SVC(probability=True)
    models['Naive Bayes'] = GaussianNB()
    models['Bagging'] = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=10, random_state=1)
    models['Boosting'] = AdaBoostClassifier(n_estimators=50, random_state=1)
    models['Stacking'] = get_stacking()
    return models

# Fungsi untuk evaluasi model menggunakan cross-validation
def evaluate_model(model, X, y):
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
    return scores

# Mendapatkan daftar file CSV
file_paths = glob.glob('C:/Users/Yessy Arye Yustraini/MSIB/datasets/*.csv')  # Ganti dengan folder tempat file Anda berada

# Iterasi untuk setiap file CSV
results_all = {}  # Untuk menyimpan hasil dari semua file
best_models = []  # Untuk menyimpan model terbaik di akhir analisis

for file_path in file_paths:
    print(f"Processing file: {file_path}")
    
    # Membaca data dari CSV dengan pemisah yang terdeteksi
    data = read_csv_with_detected_delimiter(file_path)
    
    # Tangani nilai hilang atau tidak valid
    data.replace('?', pd.NA, inplace=True)  # Ganti '?' dengan NaN
    data = data.apply(pd.to_numeric, errors='coerce')  # Konversi semua kolom ke numerik

    # Pisahkan fitur dan target
    X = data.iloc[:, :-1].values  # Semua kolom kecuali yang terakhir
    y = data.iloc[:, -1].values   # Kolom terakhir sebagai target

    # Imputasi nilai NaN
    imputer = SimpleImputer(strategy='mean')  # Bisa juga menggunakan 'median' atau 'most_frequent'
    X = imputer.fit_transform(X)
    
    # Mendapatkan model ensemble
    models = get_models()
    
    # Evaluasi setiap model
    results, names, accuracies = [], [], []
    for name, model in models.items():
        scores = evaluate_model(model, X, y)
        results.append(scores)
        names.append(name)
        accuracies.append(mean(scores))  # Simpan rata-rata akurasi
        print('>%s %.3f (%.3f)' % (name, mean(scores), std(scores)))
    
    # Simpan hasil evaluasi untuk dataset ini
    results_all[file_path] = {"names": names, "results": results}

    # Menentukan model dengan akurasi tertinggi
    best_idx = accuracies.index(max(accuracies))
    best_models.append({
        "dataset": file_path.split('/')[-1],
        "model": names[best_idx],
        "accuracy": accuracies[best_idx]
    })

    # Plot model performance untuk file ini
    plt.figure()
    plt.boxplot(results, labels=names, showmeans=True)
    plt.title(f"Model Performance Comparison ({file_path.split('/')[-1]})")
    plt.ylabel("Accuracy")
    plt.show()

# Menampilkan hasil evaluasi dan kesimpulan
print("\nKesimpulan Metode Terbaik:")
for best in best_models:
    print(f"Dataset: {best['dataset']}, Best Model: {best['model']}, Accuracy: {best['accuracy']:.3f}")
```

```python
# import libraries
import matplotlib.pyplot as plt #u/ membuat grafik
import matplotlib.ticker as ticker # u/ membuat grafik
import pandas as pd # memanggil libraries pandas dan emmberi nama alias pd
```

```python
# proses data cleaning
# menghapus kolom row ID karena bukan bagian data dengan perintah drop 
df.drop('Row ID', axis=1, inplace = True)

# axis = 1 artinya satu kolom penuh, inplace =true artinya hapus permanen
```

```python
# konversi tanggal order dengan tanggal format pandas

df['Order Date']=pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
# convert shipping date to pandas datetime format
df['Ship Date']=pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
df

# nb. date harus diubah ke format pandas karena supaya bisa di urutkan
```

```python
# sorting data by order date sekaligus dibuat permanen dengan urutan inplace=true
df.sort_values(by=['Order Date'], inplace = True, ascending=True)
```

```python
# menampilkan kolom tertentu di df 
# sama dengan : select 'Order ID','Order Date', 'Ship Date','Ship Mode'
# from df
df.filter(['Order ID','Order Date', 'Ship Date','Ship Mode'])
```

```python
#Menampilkan data sales dari customer bernama Darren Powers
df.loc[(df['Customer Name'] == 'Darren Powers')]
```

```python
# menampilkan data sales dari customer bernama Darren Powers
# dan ship mode nya standard class
df.loc[((df['Customer Name'] == 'Darren Powers')
       &(df['Ship Mode']== 'Standard Class'))]
```

```python
#Menampilkan data beberapa customer yang diinginkan
teman = ["Patrick O'Donnell", 'Darren Powers', 'Sule']
df.loc[df['Customer Name'].isin(teman)]
```

```python
#menampilkan tabel sales yang kota-nya di houston
# tapi segment nya bukan consumer 
df.loc[((df['City'] == 'Huston')&(df['Segment'] != 'Consumer'))]
```

```python
#Melihat data unik di kolom kategorial
#Melihay data unik di kolom ship mode
print(df['Ship Mode'].unique())

#Melihat daa unik di kolom categori
print(df['Category'].unique())

#Melihat data unik di kolom subCategory
print(df['Sub-Category'].unique())
```

```python
#Analisis 1
#melihat jumlah rekap penjualan 10 store urut dari yang terkecil

stateterkeci = df.filter(['State','Sales'])
stateterkeci.groupby(['State']).sum()['Sales'].nsmallest(10)
```

```python
#Analsis 2
#Membuat grafik rekap sales per sub category barang
plt.rcParams['figure.figsize']=(20,3)
plt.bar(df['Sub-Category'], df['Sales'])
plt.rcParams.update({'font.size':10})
plt.xlabel('Sub-Category')
plt.ylabel('Sales')
```

```python
#Analisis 3
#Menampilkan 20 orang customer yang paling banyak keluar uang di perusahaan

#group_by-sum = grup data per orang sekaligus salesnya dijumlah
Top_cust = df[['Customer Name', 'Sales']].groupby(["Customer Name"]).sum()


#sort+values : mengurutkan berdasarkan kolom tertentu
#ascending = False : urut dari besar ke kecil
Top_cust = Top_cust.sort_values("Sales", ascending=False)

#head(20) : menga,bil 20 data orang pertama dari atas
Top_cust = Top_cust.head(20)

#formatting nilai dari sales lebih rapi hanya dengan 2digt belakang koma
Top_cust = Top_cust[['Sales']].round(2)

# memberi indx pada top_cust
Top_cust.reset_index(inplace=True)
Top_cust
```

```python
#hanya bisa dijalankan setelah top_cust diproses
plt.figure(figsize=(15,5)) #panjang = 15 tinggi 5

#mengambar grafik denganwarna green
#menuliskan judul grafik dengan f size 18
plt.title("Most valueable customer 2015- 2019", fontsize=18)
plt.bar(Top_cust['Customer Name'],
       Top_cust['Sales'],color='#99ff99', edgecolor='green',
       linewidth=1)
plt.xlabel("Customers", fontsize=15)
plt.ylabel("Revenue", fontsize=15)
plt.xticks(fontsize=12, rotation=90)
plt.yticks(fontsize=12)
#mencetak setiap angka dalam batang grafik
for k,v in Top_cust["Sales"].items():
    plt.text(k,v-8000, '$'+str(v), fontsize=12, rotation=90,color='k',
             horizontalalignment='center')
```

```python
#Analisis 3
#Menampilkan 10 city yang paling banyak sales 

#group_by-sum = grup data per city sekaligus salesnya dijumlah
Top_city = df[['City', 'Sales']].groupby(["City"]).sum()


#sort+values : mengurutkan berdasarkan kolom tertentu
#ascending = False : urut dari besar ke kecil
Top_city = Top_city.sort_values("Sales", ascending=False)

#head(20) : mengambil 10 data city pertama dari atas
Top_city = Top_city.head(10)

#formatting nilai dari sales lebih rapi hanya dengan 2digt belakang koma
Top_city = Top_city[['Sales']].round(2)

# memberi indx pada top_city
Top_city.reset_index(inplace=True)
Top_city
```

##### Soal no 3
Tabel & Grafik yang menampilkan Sales 10 Sub Category Barang Terbanyak

```python
# Mengelompokkan data berdasarkan Sub-Category dan menghitung total Sales
Top_subcategories = df[['Sales', 'Sub-Category']].groupby(['Sub-Category']).sum()

# Mengurutkan dari besar ke kecil
Top_subcategories = Top_subcategories.sort_values("Sales", ascending=False)

# Mengambil 10 sub-kategori dengan penjualan tertinggi
Top_subcategories = Top_subcategories.head(10)

# Merapikan format angka menjadi 2 digit desimal
Top_subcategories = Top_subcategories.round(2)

# Menambahkan indeks baru
Top_subcategories.reset_index(inplace=True)

# Menampilkan tabel
Top_subcategories
```

```python
# Membuat grafik batang
plt.figure(figsize=(15, 5))  # panjang=15, tinggi=5
plt.bar(Top_subcategories['Sub-Category'], 
        Top_subcategories['Sales'], 
        color='#99ff99', edgecolor='green', linewidth=1)

# Menambahkan judul dan label
plt.title("Top 10 Sub-Category with Highest Sales", fontsize=18)
plt.xlabel("Sub-Category", fontsize=15)
plt.ylabel("Sales", fontsize=15)
plt.xticks(fontsize=12, rotation=90)
plt.yticks(fontsize=12)

# Menambahkan label angka pada batang
for k, v in enumerate(Top_subcategories["Sales"]):
    plt.text(k, v - 0.05 * v, f'${v:,.2f}', fontsize=12, rotation=90, 
             color='black', horizontalalignment='center')

# Menampilkan grafik
plt.show()
```
